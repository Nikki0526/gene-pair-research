```python
#clean and merge gene expression level, gene name and PFS data
import csv
import os
import numpy as np
import pandas as pd

data = pd.read_csv('TCGA-matrix-revised.csv')
data
```

```python
'''delete NAN'''
data = data.T
```

```python
data.ix['hgnc_symbol']=data.ix['hgnc_symbol'].fillna('null')
cols=[x for i,x in enumerate(data.columns) if data.iat[382,i]=='null']
```

```python
data=data.drop(cols,axis=1)
```

```python
data = data.T
tmp = data[['ensemble_id','hgnc_symbol']]
tmp.columns = ['Unnamed: 0','hgnc_symbol']
tmp
```

```python
data_normalized = pd.read_csv('TCGA_Normalized.csv')
data_normalized
```

```python
data1 = pd.merge(data_normalized, tmp, on = 'Unnamed: 0')
```

```python
data = data1.T
```

```python
gene_name = list(data.ix['hgnc_symbol'])
data.columns = gene_name
```

```python
data['sample'] = data._stat_axis.values.tolist()
```

```python
data2 = data.drop(data.tail(1).index)
data2 = data2.drop('Unnamed: 0',axis=0,inplace=False)
```

```python
data2['break1'] = data2['sample'].map(lambda x:x.split('-')[0])
data2['break2'] = data2['sample'].map(lambda x:x.split('-')[1])
data2['break3'] = data2['sample'].map(lambda x:x.split('-')[2])
```

```python
data2['sample_id']=data2["break1"] +"-"+ data2["break2"] +"-"+ data2["break3"]
```

```python
pfs = pd.read_csv('PFS10.csv')
```

```python
pfs.columns = ['sample_id','label_1','PFS']
```

```python
#pfs['label'] = pfs.PFS.apply(lambda x: 1 if x > 182 else 0)
pfs['label'] = pfs['label_1']
```

```python
data3 = pd.merge(pfs, data2, on = 'sample_id')
```

```python
data3.to_csv('check10.csv')
```
